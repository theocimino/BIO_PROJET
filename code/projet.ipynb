### **Question 2 - Adimensionnalisation**

On consid√®re le syst√®me initial:

\[
\begin{cases}
\dot{P} = r(H)P - \delta P^2 \\
\dot{H} = \alpha P H - m H
\end{cases}
\]

On cherche des changements de variables pour obtenir un syst√®me de la forme :

$$
\begin{cases}
\dot{P} = r(H)P - \delta P^2 \\
\dot{H} = P H - H
\end{cases}
$$

On pose les variables sans dimension :

$$
\begin{aligned}
&x = k_1P, \qquad y = k_2H, \qquad \tau = k_3 t .
\end{aligned}
$$

**Calcul de $\dfrac{dx}{d\tau}$**

- D√©veloppement :

$$
\begin{aligned}
&\frac{dx}{d\tau}
= \frac{d(k_1P)}{d(k_3t)}
= \frac{k_1}{k_3}\frac{dP}{dt} \\
&= \frac{k_1}{k_3}(r(H)P-\delta P^2) \\
&= \frac{k_1}{k_3}\Bigl[r\!\Bigl(\frac{y}{k_2}\Bigr)\frac{x}{k_1}
- \delta\Bigl(\frac{x}{k_1}\Bigr)^2\Bigr] \\
&= \frac{1}{k_3}\Bigl[r\!\Bigl(\frac{y}{k_2}\Bigr)\,x
- \Bigl(\frac{\delta}{k_1}\Bigr) x^2\Bigr].
\end{aligned}
$$

- Donc :

$$
\begin{aligned}
&\frac{dx}{d\tau} = \frac{1}{k_3}\Bigl[r\!\Bigl(\frac{y}{k_2}\Bigr)\,x
- \Bigl(\frac{\delta}{k_1}\Bigr) x^2\Bigr].
\end{aligned}
$$

**Calcul de $\dfrac{dy}{d\tau}$**

- D√©veloppement :

$$
\begin{aligned}
&\frac{dy}{d\tau}
= \frac{d(k_2H)}{d(k_3t)}
= \frac{k_2}{k_3}\frac{dH}{dt} \\
&= \frac{k_2}{k_3}(\alpha PH-mH) \\
&= \frac{k_2}{k_3}\Bigl[\alpha \frac{x}{k_1}\frac{y}{k_2}
- m\frac{y}{k_2}\Bigr] \\
&= \frac{1}{k_3}\Bigl[\frac{\alpha}{k_1}xy-my\Bigr].
\end{aligned}
$$

On pose $k_3 = m$ et $k_1 = \alpha/m$, alors :

$$
\begin{aligned}
&\frac{dy}{d\tau}
= \frac{1}{m}\Bigl[\frac{\alpha}{\alpha/m}xy-my\Bigr]
= xy-y.
\end{aligned}
$$

On r√©injecte $k_3=m$ et $k_1=\alpha/m$ dans l‚Äôexpression de $\dfrac{dx}{d\tau}$ :

$$
\begin{aligned}
&\frac{dx}{d\tau}
= \frac{1}{m}\Bigl[r\!\Bigl(\frac{y}{k_2}\Bigr)\,x
- \frac{\delta}{\alpha/m}x^2\Bigr] \\
&= \frac{1}{m}r\!\Bigl(\frac{y}{k_2}\Bigr)\,x
- \frac{\delta}{\alpha}x^2.
\end{aligned}
$$
On peut conserver $\delta = \frac{\delta}{\alpha}$ sans cons√©quence car $\alpha$ positif et $\delta$ non fix√©.
En posant $r_1(z)=\frac{1}{m}r(z)$ et $k_2=1$, on obtient :

$$
\begin{aligned}
&\frac{dx}{d\tau} = r_1(y)x - \delta x^2, \\
&\frac{dy}{d\tau} = xy - y.
\end{aligned}
$$

Finalement, en renommant les variables, le syst√®me adimensionnalis√© est de la forme :

$$
\begin{cases}
\dot{P} = r_1(H)P - \delta P^2, \\
\dot{H} = PH - H,
\end{cases}
$$

avec $r_1(H) = \frac{1}{m}r(H)$.
