### **Question 2 - Adimensionnalisation**

On consid√®re le syst√®me initial:

\[
\begin{cases}
\dot{P} = r(H)P - \delta P^2 \\
\dot{H} = \alpha P H - m H
\end{cases}
\]

On cherche des changements de variables pour obtenir un syst√®me de la forme :

$$
\begin{cases}
\dot{P} = r(H)P - \delta P^2 \\
\dot{H} = P H - H
\end{cases}
$$

On pose les variables sans dimension :

$$
\begin{aligned}
&x = k_1P, \qquad y = k_2H, \qquad \tau = k_3 t .
\end{aligned}
$$

**Calcul de $\dfrac{dx}{d\tau}$**

- D√©veloppement :

$$
\begin{aligned}
&\frac{dx}{d\tau}
= \frac{d(k_1P)}{d(k_3t)}
= \frac{k_1}{k_3}\frac{dP}{dt} \\
&= \frac{k_1}{k_3}(r(H)P-\delta P^2) \\
&= \frac{k_1}{k_3}\Bigl[r\!\Bigl(\frac{y}{k_2}\Bigr)\frac{x}{k_1}
- \delta\Bigl(\frac{x}{k_1}\Bigr)^2\Bigr] \\
&= \frac{1}{k_3}\Bigl[r\!\Bigl(\frac{y}{k_2}\Bigr)\,x
- \Bigl(\frac{\delta}{k_1}\Bigr) x^2\Bigr].
\end{aligned}
$$

- Donc :

$$
\begin{aligned}
&\frac{dx}{d\tau} = \frac{1}{k_3}\Bigl[r\!\Bigl(\frac{y}{k_2}\Bigr)\,x
- \Bigl(\frac{\delta}{k_1}\Bigr) x^2\Bigr].
\end{aligned}
$$

**Calcul de $\dfrac{dy}{d\tau}$**

- D√©veloppement :

$$
\begin{aligned}
&\frac{dy}{d\tau}
= \frac{d(k_2H)}{d(k_3t)}
= \frac{k_2}{k_3}\frac{dH}{dt} \\
&= \frac{k_2}{k_3}(\alpha PH-mH) \\
&= \frac{k_2}{k_3}\Bigl[\alpha \frac{x}{k_1}\frac{y}{k_2}
- m\frac{y}{k_2}\Bigr] \\
&= \frac{1}{k_3}\Bigl[\frac{\alpha}{k_1}xy-my\Bigr].
\end{aligned}
$$

On pose $k_3 = m$ et $k_1 = \alpha/m$, alors :

$$
\begin{aligned}
&\frac{dy}{d\tau}
= \frac{1}{m}\Bigl[\frac{\alpha}{\alpha/m}xy-my\Bigr]
= xy-y.
\end{aligned}
$$

On r√©injecte $k_3=m$ et $k_1=\alpha/m$ dans l‚Äôexpression de $\dfrac{dx}{d\tau}$ :

$$
\begin{aligned}
&\frac{dx}{d\tau}
= \frac{1}{m}\Bigl[r\!\Bigl(\frac{y}{k_2}\Bigr)\,x
- \frac{\delta}{\alpha/m}x^2\Bigr] \\
&= \frac{1}{m}r\!\Bigl(\frac{y}{k_2}\Bigr)\,x
- \frac{\delta}{\alpha}x^2.
\end{aligned}
$$
On peut conserver $\delta = \frac{\delta}{\alpha}$ sans cons√©quence car $\alpha$ positif et $\delta$ non fix√©.
En posant $r_1(z)=\frac{1}{m}r(z)$ et $k_2=1$, on obtient :

$$
\begin{aligned}
&\frac{dx}{d\tau} = r_1(y)x - \delta x^2, \\
&\frac{dy}{d\tau} = xy - y.
\end{aligned}
$$

Finalement, en renommant les variables, le syst√®me adimensionnalis√© est de la forme :

$$
\begin{cases}
\dot{P} = r_1(H)P - \delta P^2, \\
\dot{H} = PH - H,
\end{cases}
$$

avec $r_1(H) = \frac{1}{m}r(H)$.


```julia
#D√©finition des variables,fonctions pour plus tard

#CONSTANTES
r0 = 1
Œ≤ = 1/2
Œ≥ = 1

function r(H)
   return r0 * ( 1 + H / Œ≤ ) * ( 1 - H / Œ≥ )
end

function model_adim(P,H, Œ¥=1)
    return [r(H) * P - Œ¥ * P^2 ; P * H - H]
end
```

### **Question 3 - Isoclines nulles**

Regardons les isoclines nulles de
$$
\begin{cases}
\dot{P} = r(H)P - \delta P^2 \\
\dot{H} = PH - H
\end{cases}
$$

**Isocline $\dot{P} = 0$** :
$$
r(H)P - \delta P^2 = 0 \quad \Rightarrow \quad P \bigl( r(H) - \delta P \bigr) = 0
$$
- Soit $P = 0$ 
- Soit $P = \frac{r(H)}{\delta}$ 

```julia
H_range = 0:0.01:2

####################################
# 1) Fen√™tre pour l'isocline P' = 0
####################################
iso_P1 = [(h, r(h)) for h in H_range if r(h) ‚â• 0]   # P = r(H)
iso_P2 = [(h, 0.0)   for h in H_range]              # P = 0

pP = plot(
    xlabel = "H", ylabel = "P",
    title = "Isoclines P' = 0",
    xlims = (0, 2), ylims = (0, 2.5),
    legend = false,
    grid = true
)

# P = r(H)
plot!(pP,
    first.(iso_P1), last.(iso_P1),
    color = :blue, linewidth = 3,
)

# P = 0
plot!(pP,
    first.(iso_P2), last.(iso_P2),
    color = :blue, linewidth = 3,
)

display(pP)
```

**Isocline $\dot{H} = 0$** :
$$
PH - H = 0 \quad \Rightarrow \quad H(P - 1) = 0
$$
- Soit $H = 0$ 
- Soit $P = 1$ 

```julia
P_range = 0:0.01:2

####################################
# 2) Fen√™tre pour les isoclines H' = 0
####################################
iso_H1 = [(p, 1.0) for p in P_range]    # H = 0
iso_H2 = [(0, p) for p in P_range]   # P = 1

pH = plot(
    xlabel = "H", ylabel = "P",
    title = "Isoclines H' = 0",
    legend = false, grid = true,
    xlims = (0, 2), ylims = (0, 2),
)

# H = 0 (axe horizontal)
plot!(pH,
    first.(iso_H1), last.(iso_H1),
    color = :red, linewidth = 3,
)

# P = 1 (droite verticale)
plot!(pH,
    first.(iso_H2), last.(iso_H2),
    color = :red, linewidth = 3,
)

display(pH)
```

Il y a 3 configurations d'√©quilibres 

```julia
using Plots

# Ranges pour le trac√©
H_range = 0:0.01:1.2
P_range = 0:0.01:2.5

# Isoclines H' = 0 (Rouges)
iso_H1 = [(p, 1.0) for p in P_range]    # P = 1
iso_H2 = [(0.0, p) for p in P_range]    # H = 0

iso_P2 = [(h, 0.0) for h in H_range]    # P = 0

# 4 Valeurs de Œ¥ (le max de r(H) est rm = 1.125)
delta_cases = [
    1.5,    # Cas 1 : r_m/Œ¥ < 1 (0 √©quilibre interne)
    1.125,  # Cas 2 (Critique) : r_m/Œ¥ = 1 (tangent, 1 √©q. double)
    1.05,   # Cas 3 : r_0/Œ¥ < 1 < r_m/Œ¥ (2 √©q. int.)
    0.8     # Cas 4 : 1 < r_0/Œ¥ (1 √©q. int.)
]

titles = [
    "Cas 1", 
    "Cas 2", 
    "Cas 3", 
    "Cas 4"
]

for (i, Œ¥) in enumerate(delta_cases)
    
    p = plot(
        xlabel = "H", ylabel = "P",
        title = titles[i],
        xlims = (0, 1.1), ylims = (0, 1.5),
        legend = false,
        grid = true
    )
    
    iso_P1 = [(h, r(h)/Œ¥) for h in H_range if r(h) ‚â• 0] # P = r(H)/Œ¥

    # Trace Isoclines P' = 0
    plot!(p, first.(iso_P1), last.(iso_P1), color = :blue, linewidth = 3)
    plot!(p, first.(iso_P2), last.(iso_P2), color = :blue, linewidth = 3)
    
    # Trace Isoclines H' = 0
    plot!(p, first.(iso_H1), last.(iso_H1), color = :red, linewidth = 3)
    plot!(p, first.(iso_H2), last.(iso_H2), color = :red, linewidth = 3)
    
    display(p)
end
```
