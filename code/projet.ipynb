## ***Question 1 :*** 
### ***a/***
r(H) appara√Æt dans le mod√®le comme une fonction de H (pop d'herbivores). Multipli√©e par P elle constitue l'influence nette de la population d'herbivores sur la croissance des plantes. Elle constitue donc un facteur d√©terminant dans la croissance de la population de plantes. Il faut consid√©rer r(H) en proportion avec P car son influence est moduler par la valeur de P, r est donc en quelques sortes une valeur pour une population de plantes par d√©faut.
Nous allons distinguer les diff√©rentes caract√©ristiques de r(H) et les expliquer une par une.

- r(0) est positive car pas d'herbivores = pas de pr√©dateur pour les plantes, les ocnditions de d√©veloppement sont donc bonnes pour les plantes
- r est croissante dans un 1er temps car selon l'√©nonc√©, une population d'herbivores de taille mod√©r√©e devant la population de plantes peut avoir un effet positif sur la croissance de ces derni√®res.
- r atteint un pic pour une proportion dite optimale entre les 2 populations atteinte √† r(H_m)
- r commence √† d√©cro√Ætre apr√®s ce pic, car en proportion, les pr√©dateurs deviennent trop nombreux par rapport aux proies.
- r chute ensuite car les effets n√©fastes de la population d'herbivores grandissent avec celle-ci

Au final, r prend la forme d'une parabole dot√© d'un coefficient hyperbololique n√©gatif avec une seule racine positive.

```julia
#c'est juste pour g√©n√©rer l'image du graph comme sur l'√©nonc√©

using Plots
using Plots: mm, px   # importe les unit√©s [web:40]

# D√©finition de la fonction r(H)
Œ≥ = 1
r(H; r0=1, Œ≤=1/6, Œ≥=1) = r0 * (1 + H/Œ≤) * (1 - H/Œ≥)

# Domaine
H = range(0, 1.1, length=200)
r_values = r.(H)
_, idx_max = findmax(r_values)   # on trouve o√π r(H) est maximal [web:30][web:39]
H_m = H[idx_max]                 # ceci est l‚Äôant√©c√©dent du maximum, i.e. H tel que r(H) est max

# Trac√©
plot(H, r_values,  
     left_margin   = 20px,
     right_margin  = 10px,
     top_margin    = 10px,
     bottom_margin = 10px,
     label="r(H)", 
     xlabel="H", 
     ylabel="r(H)", 
     xformatter = _ -> "",   # enl√®ve les nombres sur l‚Äôaxe x [web:15][web:6]
     #yformatter = _ -> "",   # enl√®ve les nombres sur l‚Äôaxe y [web:15][web:6]
     lw=2, 
     legend = false, 
     color=:black,
     grid = false)

# Ajout de lignes de rep√®re pour Hm et max(r(H))

annotate!(H_m, minimum(r_values)-0.4, text("Hm", :bottom))
annotate!(-0.1, maximum(r_values), text("max(r(H))", :bottom))
annotate!(-0.1, r_values[1], text("r(0)", :bottom))
vline!([H_m], color=:gray, linestyle=:dash)
hline!([maximum(r_values)], color=:gray, linestyle=:dash)
title_plot = ""

# Mise en forme
plot!(title = title_plot)
```
